```python
import csv
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

path = r"C:\SandBox\Prosperity\Data1"  # Use a raw string
os.chdir(path)
print(os.getcwd())

data = "activity.csv"
```

```python
data = pd.read_csv(data)
data.head()
```

```python
amethystData = data[data['product'] == 'AMETHYSTS']
starfruitData = data[data['product'] == 'STARFRUIT']

amethystData.head()
```

```python
# Plot Amethyst midprice
plt.figure(figsize=(8, 3))
plt.plot(amethystData['timestamp'], amethystData['mid_price'], color='r')
plt.ylim(9995, 10005)
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Amethyst Price')
plt.show()

# Plot Starfruit midprice
plt.figure(figsize=(8, 3))
plt.plot(starfruitData['timestamp'], starfruitData['mid_price'] , color='g')
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Starfruit Price')
plt.show()
```
