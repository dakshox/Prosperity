```python
import tools.log_parser as log_parser
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style="whitegrid")
```

```python
LOG_PATH = "../data/prober.log"
parsed_logs = log_parser.parse_log(LOG_PATH)
```

```python

# sns.lineplot(parsed_logs.activity_df, x="timestamp", y="mid_price", errorbar=None)
prices = sns.relplot(
    data=parsed_logs.activity_df,
    x="timestamp",
    y="mid_price",
    col="product",
    kind="line",
    col_wrap=2,
    legend=False,
    facet_kws={"sharey": False},
)

for product, ax in prices.axes_dict.items():
    ax.set_title(product)
```
