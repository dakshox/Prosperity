```python
# Adjusting the delimiter to semicolon and reloading the data
data = pd.read_csv('/Users/yoonsanglee/Documents/GitHub/Prosperity/Daksh.csv', delimiter=';')

# Display the first few rows of the dataframe with the correct delimiter
data.head()
```

```python
import matplotlib.pyplot as plt

# Filtering data for AMETHYSTS and STARFRUIT
amethysts = data[data['product'] == 'AMETHYSTS']
starfruit = data[data['product'] == 'STARFRUIT']

# Time Series plot of Mid Prices
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(amethysts['timestamp'], amethysts['mid_price'], label='AMETHYSTS Mid Price')
plt.xlabel('Timestamp')
plt.ylabel('Mid Price')
plt.title('AMETHYSTS Mid Price Over Time')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(starfruit['timestamp'], starfruit['mid_price'], label='STARFRUIT Mid Price', color='orange')
plt.xlabel('Timestamp')
plt.ylabel('Mid Price')
plt.title('STARFRUIT Mid Price Over Time')
plt.legend()

plt.tight_layout()
plt.show()

# Histogram of Profit and Loss for both AMETHYSTS and STARFRUIT
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.hist(amethysts['profit_and_loss'], bins=20, color='blue', alpha=0.7)
plt.xlabel('Profit and Loss')
plt.ylabel('Frequency')
plt.title('AMETHYSTS Profit and Loss Distribution')

plt.subplot(1, 2, 2)
plt.hist(starfruit['profit_and_loss'], bins=20, color='orange', alpha=0.7)
plt.xlabel('Profit and Loss')
plt.ylabel('Frequency')
plt.title('STARFRUIT Profit and Loss Distribution')

plt.tight_layout()
plt.show()
```
