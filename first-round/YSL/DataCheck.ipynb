```python
import pandas as pd 

file_path = '/Users/yoonsanglee/Documents/GitHub/Prosperity/tutorial-round/YSL-work/TradingData-given.csv'

results_df = pd.read_csv(file_path)

results_df.head()
```

```python
# Reload the CSV file with the correct delimiter
data_corrected = pd.read_csv(file_path, delimiter=';')

# Display the first few rows of the dataframe with the correct structure
data_corrected.head()
```

```python
import matplotlib.pyplot as plt

# Filter data for the "AMETHYSTS"
amethysts_data = data_corrected[data_corrected['product'] == 'AMETHYSTS']

plt.figure(figsize=(14, 7))
plt.plot(amethysts_data['timestamp'], amethysts_data['bid_price_1'], label='Bid Price', color='blue', marker='o')
plt.plot(amethysts_data['timestamp'], amethysts_data['ask_price_1'], label='Ask Price', color='red', marker='x')

plt.title('Bid and Ask Prices Over Time for AMETHYSTS')
plt.xlabel('Timestamp')
plt.ylabel('Price')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Filter data for the "STARFRUIT"
starfruit_data = data_corrected[data_corrected['product'] == 'STARFRUIT']

plt.figure(figsize=(14, 7))
plt.plot(starfruit_data['timestamp'], starfruit_data['bid_price_1'], label='Bid Price', color='blue', marker='o')
plt.plot(starfruit_data['timestamp'], starfruit_data['ask_price_1'], label='Ask Price', color='red', marker='x')

plt.title('Bid and Ask Prices Over Time for STARFRUIT')
plt.xlabel('Timestamp')
plt.ylabel('Price')
plt.legend()
plt.grid(True)

plt.show()
```

```python
trading_summary = data_corrected.groupby('timestamp')['profit_and_loss'].sum().reset_index()

plt.figure(figsize=(14, 7))
plt.plot(trading_summary['timestamp'], trading_summary['profit_and_loss'], label='Profit and Loss', color='green', marker='^')

plt.title('Overall Trading Profit and Loss Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Profit and Loss')
plt.legend()
plt.grid(True)

plt.show()

trading_summary.head()
```
