```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('C:/SandBox/Prosperity/data/raw/prices_round_1_day_0.csv', delimiter=';')
data.head()
```

```python
starfruit = data[data['product'] == 'STARFRUIT']
starfruit.head()
```

```python
starfruit['future_price'] = starfruit['mid_price'].shift(-1)
starfruit.head()
```

```python
starfruit['ewma1'] = starfruit['mid_price'].ewm(alpha=0.1).mean()
starfruit['ewma01'] = starfruit['mid_price'].ewm(alpha=0.01).mean()
```
