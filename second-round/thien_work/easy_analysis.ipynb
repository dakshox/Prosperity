```python
import tools.log_parser as log_parser
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style="whitegrid")
```

```python
LOG_PATH = "../../data/data_v1"
```

```python
r1_price_df = pd.read_csv(f"{LOG_PATH}/prices_1.csv", sep=";")
r2_price_df = pd.read_csv(f"{LOG_PATH}/prices_2.csv", sep=";")
```

```python

# sns.lineplot(parsed_logs.activity_df, x="timestamp", y="mid_price", errorbar=None)
prices = sns.relplot(
    data=r1_price_df,
    x="total_timestamp",
    y="mid_price",
    col="product",
    kind="line",
    col_wrap=2,
    legend=False,
    facet_kws={"sharey": False},
)

for product, ax in prices.axes_dict.items():
    ax.set_title(product)
```

```python
# r2_price_df.plot(x="total_timestamp", y=["ORCHIDS", "SUNLIGHT"])
r2_plot_df = r2_price_df.set_index("total_timestamp", drop=True, inplace=False)
categories = ["ORCHIDS", "TRANSPORT_FEES", "IMPORT_TARIFF", "EXPORT_TARIFF", "SUNLIGHT", "HUMIDITY"]
fig, axs = plt.subplots(len(categories), 1, figsize=(10, 5 * len(categories)))
for i, category in enumerate(categories):
    sns.lineplot(data=r2_plot_df, x="total_timestamp", y=category, ax=axs[i], errorbar=None)
    if i != 0:
        ax2 = axs[i].twinx()
        ax2.grid(False)
        sns.lineplot(data=r2_plot_df, x="total_timestamp", y="ORCHIDS", ax=ax2, legend=False, errorbar=None, alpha=0.4)
```
