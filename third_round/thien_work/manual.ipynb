```python
MULTS = [24, 70, 41, 21, 60,
         47, 82, 87, 80, 35,
         73, 89, 100, 90, 17,
         77, 83, 85, 79, 55,
         12, 27, 52, 15, 30]
HUNTS = [2, 4, 3, 2, 4,
         3, 5, 5, 5, 3,
         4, 5, 8, 7, 2,
         5, 5, 5, 5, 4,
         2, 3, 4, 2, 3]
```

```python
import matplotlib.pyplot as plt
```

```python


def display(aux):
    # Display in a 5x5 grid without axes
    # turn off axes
    plt.figure(figsize=(1, 1))
    plt.axis('off')
    # use text
    for i in range(25):
        plt.text(i % 5, 4 - i // 5 + 0.2, f"{aux[i]:.2f}", ha='center', va='center', color='blue')
        plt.text(i % 5, 4 - i // 5, str(MULTS[i]), ha='center', va='center')
        plt.text(i % 5, 4 - i // 5 - 0.2, str(HUNTS[i]), ha='center', va='center', color='red')

display([0.0] * 25)
```

```python
def required(payout):
    result = []
    for i in range(25):
        # payout = mult / (hunt + res)
        # res = mult / payout - hunt
        res = min(100.0, MULTS[i] / payout - HUNTS[i])
        if res >= 0:
            result.append(res)
        else:
            result.append(0.0)
    return result
```

```python
display(required(1.5))
```

```python
ans = 0.00
add = 32
for _ in range(30):
    cand = ans + add
    s = sum(required(cand))
    if s > 100:
        ans = cand
    add /= 2
print(ans)
display(required(ans))
```
