```python
import pandas as pd
import matplotlib.pyplot as plt
data_filepath = "results-03-19-01/activity.csv"
```

```python
df = pd.read_csv(data_filepath)
df.head()
```

```python
starfruit_data = df[df['product'] == 'STARFRUIT']
```

```python
starfruit_data.index = starfruit_data['timestamp']
plt.plot(starfruit_data['mid_price'])
```

```python
starfruit_data['ewma05'] = starfruit_data['mid_price'].ewm(alpha=0.05, adjust=False).mean()
starfruit_data['ewma1'] = starfruit_data['mid_price'].ewm(alpha=0.1, adjust=False).mean()
starfruit_data['ewma2'] = starfruit_data['mid_price'].ewm(alpha=0.2, adjust=False).mean()
starfruit_data['ewma3'] = starfruit_data['mid_price'].ewm(alpha=0.3, adjust=False).mean()
```

```python
plt.figure(figsize=(12, 5))
# plt.plot(starfruit_data['ewma05'], color='red', label='ewma05')
plt.plot(starfruit_data['ewma1'], color='green', label='ewma1')
# plt.plot(starfruit_data['ewma2'], color='orange', label='ewma2')
# plt.plot(starfruit_data['ewma3'], color='purple', label='ewma3')
plt.plot(starfruit_data['mid_price'], color='blue', label='mid_price', alpha = 0.3)
plt.legend()

```
