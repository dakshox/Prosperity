### Analysis of results-03-12-02 

This code works and makes money, if jump in bid/ask is greater than plus minus 4, it will buy/sell. 

```python
data5 = pd.read_csv("results-03-12-02/activity.csv")
amethystData5 = data5[data5['product'] == 'AMETHYSTS']
starfruitData5 = data5[data5['product'] == 'STARFRUIT']
starfruitData5.head()
```

```python
plt.figure(figsize=(8, 3))
plt.plot(starfruitData5['timestamp'], starfruitData5['profit_and_loss'], color='r', label = "PnL")
plt.plot(starfruitData5['timestamp'], [0 for i in range(len(starfruitData5['timestamp']))] , color='g', label = " PnL = 0", linestyle = '--')
plt.xlabel('Time')
plt.ylabel('Profit and Loss')
plt.title('Starfruit PnL')
```

Look at timeframes 75000 - 100000

```python
negPnL = starfruitData5[(starfruitData5['timestamp'] > 75000) & (starfruitData5['timestamp'] < 80000)]
negPnL.head()
```

```python
plt.figure(figsize=(8, 3))
plt.plot(negPnL['timestamp'], negPnL['profit_and_loss'], color='r', label = "PnL")
# plt.plot(negPnL['timestamp'], [0 for i in range(len(negPnL['timestamp']))] , color='g', label = " PnL = 0", linestyle = '--')
plt.xlabel('Time')
plt.ylabel('Profit and Loss')
plt.title('Starfruit PnL')
plt.show()
```

```python
# Assuming negPnL is a pandas DataFrame
negPnL = negPnL.reset_index(drop=True)
marker = pd.Series([False] * len(negPnL))  # Initialize a boolean Series with the same length as negPnL

for i in range(1, len(negPnL)):  # Start from 1 since we're comparing to the previous item
    if negPnL.bid_price_1.iloc[i] > negPnL.bid_price_1.iloc[i-1] + 4:
        marker.iloc[i] = True
        marker.iloc[i-1] = True
    elif negPnL.ask_price_1.iloc[i] < negPnL.ask_price_1.iloc[i-1] - 4:
        marker.iloc[i] = True#
        marker.iloc[i-1] = True

bigJump = negPnL[marker]  # Use the boolean Series for indexing
```

```python
plt.figure(figsize=(8, 3))
plt.plot(negPnL['timestamp'], negPnL['mid_price'], color='r', label = "mid price")
plt.scatter(negPnL['timestamp'], negPnL['bid_price_1'], color='g', label = "bid price", marker = '.')
plt.scatter(negPnL['timestamp'], negPnL['ask_price_1'], color='b', label = "ask price", marker = '.')
plt.scatter(bigJump['timestamp'], bigJump['bid_price_1'], color='g', label = "big jump")
plt.scatter(bigJump['timestamp'], bigJump['ask_price_1'], color='b', label = "big jump")
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Starfruit Price')
plt.legend()
plt.show()
plt.figure(figsize=(8, 3))
plt.plot(negPnL['timestamp'], negPnL['profit_and_loss'], color='r', label = "PnL")
# plt.plot(negPnL['timestamp'], [0 for i in range(len(negPnL['timestamp']))] , color='g', label = " PnL = 0", linestyle = '--')
plt.xlabel('Time')
plt.ylabel('Profit and Loss')
plt.title('Starfruit PnL')
plt.show() 
```
